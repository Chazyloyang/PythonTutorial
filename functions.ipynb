# Functions in Python

Functions are reusable block of code written to perform a particular task or a perticular mathematical operation. This helps to prevent writing code repeatedly. A function in python is defined using the "def" keyword

```python
#write a function to add two numbers
def addition(a,b):
    return a+b;

#calling the function addition
addition(4,7)
```

```python
#write a function to convert temperature from celcuis to fahrenheit
def fahrenheit(temp):
    return temp + 273;


#calling the function
fahrenheit(30)
```
